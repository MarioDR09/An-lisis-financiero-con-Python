## Monte Carlo (predicci√≥n de ganancia neta)

```python
import numpy as np
import matplotlib.pyplot as plt

rev_m = 200
rev_stdev = 10
it = 256

rev = np.random.normal(rev_m, rev_stdev, it)
COGS = - (rev * np.random.normal(0.4,0.2))

COGS.mean(), COGS.std()
```

****

```python
Gan_Net = rev + COGS
Gan_Net
```

```python
plt.figure(figsize=(15, 6))
plt.plot(Gan_Net)
plt.show()
```

Las ganancias m√≠nimas y m√°ximas que se obtuvieron fueron:

```python
max(Gan_Net)
```

```python
min(Gan_Net)
```

La media y desviaci√≥n est√°ndar

```python
Gan_Net.mean()
```

```python
Gan_Net.std()
```

Graficamos los datos en un histograma. Usamos 20 bins para ver mejor la distribuci√≥n de los datos.

```python
plt.figure(figsize=(10, 6));
plt.hist(Gan_Net, bins = 20);
plt.show()
```
