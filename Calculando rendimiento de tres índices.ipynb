## Calculando rendimiento de tres índices

Considerando tres índices conocidos – Dow Jones, S&P 500, y el Nasdaq desde el primero de Enero del 2000.

```python
import numpy as np
import pandas as pd
pd.core.common.is_list_like = pd.api.types.is_list_like
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
ind_datos = pd.read_csv('C:/Users/Mario/Finanzas Python/Indices_Exercise_Data.csv', index_col='Date')
```

```python
ind_datos.head()
```

```python
ind_datos.tail()
```

Normalizamos a 100 y posteriormente graficamos los resultados. 

```python
(ind_datos / ind_datos.iloc[0] * 100).plot(figsize=(15, 6));
plt.show()
```

*****

Obtenemos el redimiento simple de los índices.

```python
ind_rend = (ind_datos / ind_datos.shift(1)) - 1

ind_rend.tail()
```

Estimate the average annual return of each index.

```python
an_ind_rend = ind_rend.mean() * 250
an_ind_rend
```
